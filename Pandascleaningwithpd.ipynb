## outliers


```python
##create a dataframe with simple data
df = pd.DataFrame({'Height': [10,170, 160, 130, 190, 180, 150, 140, 200, 175, 165, 1000]})
df
```

```python
#calculate the z_score of each data point
z_scores = np.abs(stats.zscore(df['Height']))
z_scores

```

```python
z_scores = np.abs(stats.zscore(df['Height']))
print(z_scores)


threshold = 3
outliers = df[z_scores > 3]

outliers
```

```python
mask =  df["Height"] > 150
mask
```

```python
df[mask]
```

### Detecting using interquartile range

## Data points that fall below q1 - 1.5 x iqr or above q3 + 1.5 x iqr are typically considered outliers

```python
# Calculate Q1, Q3, and IQR
Q1 = df['Height'].quantile(0.25)
Q3 = df['Height'].quantile(0.75)

IQR = Q3 - Q1

threshold = 1.5

mask1 = Q1 - threshold * IQR
mask2 = Q3 + threshold * IQR

outliers = df[(df['Height'] < mask1) | (df['Height'] > mask2)]

print(outliers)
```

```python
IQR
```

## how to drop outliers

```python
outliers.index
```

```python
##drop outliers from their rows
df.drop(outliers.index, inplace=True)
```

```python
df.reset_index(inplace=True, drop=True)
```

## Encoding categorical variables

```python
# Create a sample dataframe
df = pd.DataFrame({'Priority': ['Low', 'Medium', 'High', 'Medium', 'Low']})
df                
```

```python
# Define a mapping dictionary
priority_mapping = {'Low': 1, 'Medium': 2, 'High': 3}

#use the replace function to apply the dictionary mapping
df['Priority_Encoded'] = df['Priority'].replace(priority_mapping)

##display the updated dataframe
df
```
